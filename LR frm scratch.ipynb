```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#linear regression from scratch
#for single variable
def model(x, w_final, b_final):
    f_wb = np.dot(w_final, x) + b_final
    return f_wb
```

```python
def cost_function(x, y, w, b):
    m = len(x)
    f_wb = np.dot(x, w) + b
    error = (f_wb - y) ** 2
    cost = np.sum(error) / (2 * m)
    return cost
```

```python
def compute_gradient(x, y, w, b):
    m = len(x)
    f_wb = np.dot(x, w) + b
    
    dw = (1/m) * np.sum((f_wb - y) * x)
    db = (1/m) * np.sum(f_wb - y)
    
    return dw, db
```

```python
def gradient_descent(x, y, ini_w, ini_b, alpha, iterations):
    w = ini_w
    b = ini_b
    cost_history = []
    for i in range(iterations):
        dw, db = compute_gradient(x, y, w, b)
        w = w - (alpha * dw)
        b = b - (alpha * db)
        cost = cost_function(x, y, w, b)
        cost_history.append(cost)
    return w, b, cost_history
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1, 2, 3, 4, 5, 6, 7])

w_init = 0
b_init = 0
alpha = 0.01
iterations = 100
w_final, b_final, cost_history = gradient_descent(x, y, w_init, b_init, alpha, iterations)

plt.plot(range(iterations), cost_history, label = "Cost Function", color = "b")
plt.xlabel("iterations")
plt.ylabel("Cost")
plt.title("Cost Function vs. Iteration")
plt.legend()
plt.show()
```
